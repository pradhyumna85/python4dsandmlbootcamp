```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with sns.axes_style("darkgrid"):  # for grid in new version
    sns.distplot(tips["total_bill"])
```

```python
with sns.axes_style("darkgrid"):  # for grid in new version
    sns.distplot(
        tips["total_bill"], hist_kws=dict(edgecolor="k", linewidth=1, alpha=0.3)
    )
```

```python
sns.distplot(
    tips["total_bill"], kde=False, hist_kws=dict(edgecolor="k", linewidth=1, alpha=0.3)
)
```

```python
sns.set(style="darkgrid", color_codes=True)  # set method to customize style
sns.distplot(
    tips["total_bill"],
    bins=30,
    # kde=False,
    hist_kws=dict(edgecolor="k", linewidth=1, alpha=0.3),
)
```

```python
with sns.axes_style("whitegrid"):
    sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
with sns.axes_style("whitegrid"):
    sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
```

```python
import scipy.stats as stats

with sns.axes_style("darkgrid"):
    sns.jointplot(
        x="total_bill", y="tip", data=tips, kind="reg", stat_func=stats.pearsonr
    )
```

```python
with sns.axes_style("darkgrid"):
    sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg").annotate(
        stats.pearsonr
    )
```

```python
with sns.axes_style("darkgrid"):
    sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
```

```python
# imp
pearson_coeff, pval = stats.pearsonr(tips["total_bill"], tips["tip"])

print(f"Pearson coeff: {pearson_coeff}, p-value: {pval}")
```
