```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
%%time
plt.plot(x, y)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Title")
plt.autoscale(enable=True, axis="both", tight=True) # for tight fitting axis -> ax.autoscale(tight=True)
# plt.show() not really required
```

```python
%%time
plt.subplot(1, 2, 1)
plt.plot(x, y, "r")
plt.subplot(1, 2, 2)
plt.plot(y, x, "b")
# plt.subplot(1,2,3) results in error, because only 2 plots per row
# plt.plot(x,y*2,'g') results in error
```

```python
# object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel("x label")
axes.set_ylabel("y label")
axes.set_title("Title")
axes.autoscale(tight=True)  # for tight fitting axis
```

```python
fig = plt.figure()
axes1 = fig.add_axes(
    [0.1, 0.1, 0.8, 0.8]
)  # left,bottom,weidht%,height% relative to 1 or 100%
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])  # left,bottom,width%,height%
axes1.plot(x, y)
axes1.set_title("larger plot")
axes2.plot(y, x)
axes2.set_title("smaller plot")
axes1.autoscale(enable=True, axis="both", tight=True)  # for tight fitting axis
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# axes.plot(x, y)
# for current in axes:
#     current.plot(x, y)
#     # break
axes[0].plot(x, y, "r")
axes[0].set_title("first plot")

axes[1].plot(y, x, "b")
axes[1].set_title("second plot")
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2, label="x squared")
ax.set_title("title")
ax.set_ylabel("y label")
ax.set_xlabel("x label")
ax.plot(x, x ** 3, "ro", label="x cubed")
# ax.legend(loc=0) #best fit
ax.legend(loc=(0.1, 0.1))
ax.axis("tight")
```
